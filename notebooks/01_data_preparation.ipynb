```python
import pandas as pd
import os
```

```python
df = pd.read_csv("../data/raw/clean_data.csv")
```

```python
df.head()
```

```python
df_subset = df[df['Dept'].isin([i for i in range(10)])]
```

```python
# Splitting data into Stores
store_dir = "../data/store_level_data_2/"
train_size = {}
for store in df_subset['Store'].unique():
    df_sub = df_subset[df_subset['Store'] == store]
    df_sub['Date'] = pd.to_datetime(df_sub['Date'])
    df_sub = df_sub.set_index("Date")
    path = os.path.join(store_dir,"store_" + str(store))
    os.makedirs(path)
    train = df_sub.loc[:'2012-05-01'].reset_index()
    valid = df_sub.loc['2012-05-02':'2012-08-01'].reset_index()
    test = df_sub.loc['2012-08-01':].reset_index()
    train_size[store] =  train.shape[0]
    train.to_csv(os.path.join(path, "train.csv"), index=False)
    valid.to_csv(os.path.join(path, "valid.csv"), index=False)
    test.to_csv(os.path.join(path, "test.csv"), index=False)
```
